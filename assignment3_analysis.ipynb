```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
df = pd.read_csv("Assignment #3 - Ai's version 5-4.csv")
df
```

#### Correlation between Red count and chain_percentage

```python
x = df["chain_percentage"]
y = df["red_count"]
slope, intercept, r, p, stderr = ss.linregress(x, y)
```

```python
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'
line
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=0, marker='o', label='Data points')
ax.plot(x, intercept + slope * x, label=line)
ax.set_xlabel('Chain Restaurant Percentage')
ax.set_ylabel("Grade 'Red' Count")
ax.legend(facecolor='white')
plt.title('Correlation between Red Count and Chain Restaurant Percentage')
plt.savefig('chain_red.png')
plt.show()
```

### Logistic Regression 

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.3, random_state=88)
df_train.shape, df_test.shape
```

```python
df_train.head()
```

```python
X_train = df_train[['median_income', 'chain_count', 'poverty_rate']]
y_train = df_train['red_count']

X_test = df_test[['median_income', 'chain_count', 'poverty_rate']]
y_test = df_test['red_count']
```

```python
import statsmodels.formula.api as smf
ols = smf.ols(formula='red_count ~ median_income + chain_count + poverty_rate',
                data=df_train).fit()
print(ols.summary())
```

```python
def OSR2_scikit(model, X_test, y_test, y_train):
    y_pred = model.predict(X_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(y_train))**2)                 
    return (1 - SSE/SST)
```

```python
OSR2_scikit(ols, X_test, y_test, y_train)
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
OSR2_scikit(linreg, X_test, y_test, y_train)
```

```python
# calculate Variance Inflation Factor for each explanatory variable
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

def VIF(df, columns):
    values = sm.add_constant(df[columns]).values
    num_columns = len(columns)+1
    vif = [variance_inflation_factor(values, i) for i in range(num_columns)]
    return pd.Series(vif[1:], index=columns)
```

```python
cols = ['median_income', 'chain_count', 'poverty_rate']
pd.DataFrame(VIF(df_train, cols)).rename(columns={0:'VIF score'})
```
